```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test/a', cv2.IMREAD_UNCHANGED)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height, width, _ = img.shape
left_eye = np.asarray([106, 104])
right_eye = np.asarray([131, 104])
leftH = np.asarray([109, 254])
rightH = np.asarray([252, 110])
horVec=rightH-leftH
eyeVec =left_eye-right_eye
eyeVec = (eyeVec)/np.sqrt(eyeVec[0]*eyeVec[0]+eyeVec[1]*eyeVec[1])
print(eyeVec)
horVec=(horVec)/np.sqrt(horVec[0]*horVec[0]+horVec[1]*horVec[1])
print(horVec)
costheta=np.dot(eyeVec, horVec)
print(costheta)
angleinRad=np.arccos(costheta)
angleindeg = angleinRad*(180/np.pi)
M1=cv2.getRotationMatrix2D((img.shape[1]/2, img.shape[0]/2), 180+angleindeg, 1)
im_rot = cv2.warpAffine(img, M1, (width+100, height+100))
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[10, 10])
ax1.imshow(img)
ax2.imshow(im_rot)
plt.show()
```
